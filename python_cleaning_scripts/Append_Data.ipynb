```python
# imports
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType,StructField, StringType, DoubleType, IntegerType, DateType
from pandas_profiling import ProfileReport
import pandas as pd
import os
```

```python
# Get all the files in the data directory.
csvFiles = os.listdir("../data")
```

```python
# Create the spark session and context.
spark = pyspark.sql.SparkSession.builder.getOrCreate()
sc = spark.sparkContext
```

```python
# read in the csv's and append the data into one dataframe.
df = spark.read.csv("../data/" + csvFiles[0], sep=",", header=True)
csvFiles.remove(csvFiles[0])

for file in csvFiles:
    newData = spark.read.csv("../data/" + file, sep=",", header=True)
    df = df.union(newData)
```

```python
# Show the data to verify.
df.show()
```

```python
# Verify the count
df.count()
```

```python
# Write the appended data to a csv.
df.toPandas().to_csv("../appended_data/appended_data.csv")
```

```python
profile = ProfileReport(df.toPandas(), title="Pandas Profiling Report")
profile
```
